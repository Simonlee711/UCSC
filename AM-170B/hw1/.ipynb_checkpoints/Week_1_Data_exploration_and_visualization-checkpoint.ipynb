#Week 1: Data exploration and visualization

```python
import seaborn as sb 
import matplotlib.pyplot as plt
import numpy as np 
```

### Visualizing categorical data 
Categorical data: one of the data types is discrete.  This can be represented in a number of ways, including a barplot, a box plot, a scatterplot. The following data is Titanic survival data. 

```python
titanic = sb.load_dataset("titanic")
sb.set(context='paper', font_scale=1.5, style='whitegrid')
sb.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
# Use the line below to save a high res image to disk 
# plt.savefig('./titanic.png', dpi=300) 
```

## Visualizing categorical data: the barplot vs. scatterplot

The barplot summarizes the data. The scatterplot combined with the boxplot shows the distribution of the data and richer visualization. 

```python
tips = sb.load_dataset("tips")
print(tips)
sb.catplot(x="day", y="tip", data=tips, kind='bar', hue='time')
plt.figure()
sb.catplot(x="day", y="tip", data=tips,hue='time', dodge=True)
sb.boxplot(x="day", y="tip",data=tips, color='white',hue='time')
```

## Visualizing distributions 
Using histograms, kernel density estimate plots, joint probability distributions. 

```python
# Distributions
penguins = sb.load_dataset("penguins")
print(penguins)
sb.displot(penguins, x="flipper_length_mm", hue="species")
sb.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill=True)
sb.displot(penguins, x="bill_length_mm", hue="species", kind="kde", fill=True)
sb.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm",  hue="species")
sb.pairplot(penguins, hue='species')
```

## Timeseries data 
We start by counting how many timeseries we have. 

```python
fmri = sb.load_dataset("fmri")
print (fmri)
```

```python
plt.figure()
sb.scatterplot(x="timepoint", y="signal",hue="region", style="event", data=fmri, estimator=None)
plt.figure()
sb.lineplot(x="timepoint", y="signal",hue="region", style="event", data=fmri)
```

## Subset part of the data and plot 
Take region == frontal and plot the same data 

```python
fmri_front = fmri.loc[fmri['region'] =='frontal']
print(fmri_front)
```

## Data can also be queried directly using the query command

```python
sb.lineplot(
    data=fmri.query("region == 'frontal'"),
    x="timepoint", y="signal", hue="event", units="subject",
    estimator=None, lw=1,
)
```
