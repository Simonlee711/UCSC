In genetics, gene expression is the most fundamental level at which genetic material gives rise to the phenotype, in other words, an observable trait.  One of the RNA sequencing is a technique to reveal the presence and quantity of RNA in a biological sample at a given moment. Analysis of RNA sequencing data facilitates uncovering differences in gene expression in different groups or treatments; the result of which is often used to direct treatments or drug development. # III Population dynamics: predator prey. 

The predatorprey problem refers to an ecological system in which we have two species, one of which feeds on the other. This type of system has been studied for decades and is known to exhibit interesting dynamics. We will visualize the evolution of these populations over time and with respect to each other, and explore the system's equilibrium points. 

1. Implement the following population dynamics model: 

$x^{.} = rx(1-\frac{x}{k}) - \frac{axy}{c+x}, x\geq 0 $

$ y^{.} = \frac{baxy}{c+x} -dy, y \geq 0$



The parameter values are a = 3.2, b = 0.6, c = 50, d = 0.56, k = 125, and r = 1.6.

2. Find the three equilibrium points of the system, by setting  and solving for x and y. Note one of equilibrium points is (0,0). 

3. Pick points near the equilibrium points solved in 2). (Or you can use initial conditions (x=25, y = 20), (x=1, y=1) , (x=20, y=2).

a) Plot a simulation of the two populations as a function of time.

b) Plot a simulation of the two populations plotted against each other. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import scipy

# constants
a = 3.2
b = 0.6
c = 50
d = 0.56
k = 125
r = 1.6


# ode set up
def population(X, t, a, b, c, d):
    x, y = X
    xdot = r*x * (1-(x/k)) - ((a*x*y)/(c+x))
    ydot = ((b*a*x*y)/(c+x)) - d*y
    return np.array([xdot, ydot])

x0 = 1.0
y0 = 1.0
t = np.linspace(0,30,1000)
X0 = [x0, y0]
ode = scipy.integrate.odeint(population, X0, t, args = (a, b, c, d))
x, y = ode.T

plt.figure()
plt.title("Population dynamics model (x0 = 1, y0 = 1)")
plt.plot(t, x, 'x', label = 'Predator')
plt.plot(t, y, '+', label = "Prey")
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.savefig('data/fig15.png')
plt.show()

plt.figure()
plt.title("Phase portrait (x0 = 1, y0 = 1)")
plt.plot(x, y, 'r')
plt.xlabel('Predator')
plt.ylabel('Prey')
plt.legend()
plt.savefig('data/fig16.png')
plt.show()

x0 = 25.0
y0 = 20.0
t = np.linspace(0,30,1000)
X0 = [x0, y0]
ode = scipy.integrate.odeint(population, X0, t, args = (a, b, c, d))
x2, y2 = ode.T

plt.figure()
plt.title("Population dynamics model (x0 = 25, y0 = 20)")
plt.plot(t, x2, 'x', label = 'Predator')
plt.plot(t, y2, '+', label = "Prey")
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.savefig('data/fig17.png')
plt.show()

plt.figure()
plt.title("Phase portrait (x0 = 25, y0 = 20)")
plt.plot(x2, y2, 'r')
plt.xlabel('Predator')
plt.ylabel('Prey')
plt.legend()
plt.savefig('data/fig18.png')
plt.show()


x0 = 20.0
y0 = 2.0
t = np.linspace(0,30,1000)
X0 = [x0, y0]
ode = scipy.integrate.odeint(population, X0, t, args = (a, b, c, d))
x3, y3 = ode.T

plt.figure()
plt.title("Population dynamics model (x0 = 20, y0 = 2)")
plt.plot(t, x3, 'x', label = 'Predator')
plt.plot(t, y3, '+', label = "Prey")
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.savefig('data/fig19.png')
plt.show()

plt.figure()
plt.title("Phase portrait (x0 = 20, y0 = 2)")
plt.plot(x3, y3, 'r')
plt.xlabel('Time')
plt.ylabel('Population')
plt.savefig('data/fig20.png')
plt.show()

# part 3


```
